```python
import pandas as pd
import numpy as np
```

```python
# Pandas 데이터형을 구성하는 기본 Series()
s = pd.Series([1, 3, 5, np.nan, 6, 8])
s
```

```python
# 날짜 이용
dates = pd.date_range("20130101", periods=6)
dates
```

```python
# Pandas에서 가장 많이 사용되는 데이터형 DataFrame(index와column을 지정해준다.)
df = pd.DataFrame(np.random.randn(6, 4), index=dates, columns=["A", "B", "C", "D"])
df
```

```python
# 앞부분 데이터 확인
df.head()
```

```python
# DataFrame의 기본 정보 확인(각 컬럼의 크기와 데이터형태를 확인한다.)
df.info()
```

```python
# DataFrame의 통계적 기본 정보를 확인
df.describe()
```

```python
# sort_values: 데이터 정렬
df.sort_values(by = "B",ascending = False) # B컬럼을 기준으로 내림차순 정렬!
```

```python
# 특정 컬럼만 읽기
df["A"]
```

```python
# [n:m]: n부터 m-1 까지(인덱스나 컬럼의 이름으로 slice하는 경우는 끝을 포함함)
df[0:3]
```

```python
df["20130102":"20130104"]
```

```python
# Pandas의 보편적인 slice 옵션(모든 행에 열 "A","B" 컬럼만 선택)
df.loc[:,["A", "B"]]
```

```python
df.loc["20130102",["A", "B"]]
```

```python
# loc는 컬럼 명이나 인덱스 명으로 지정, iloc는 인덱스의 번호로만 지정
df.iloc[3]
```

```python
df.iloc[3:5, 0:2]
```

```python
df.iloc[[1, 2, 4], [0, 2]]
```

```python
df.iloc[:, 1:3]
```

```python
# A컬럼의 값들 중 0보다 큰값들만 출력
df[df["A"] > 0]
```

```python
df[df>0]
```

```python
# DataFrame에 E라는 컬럼을 추가
df["E"] = ["one", "one", "two", "three", "four", "three"]
df
```

```python
# 특정 요소가 있는지 확인
df["E"].isin(["two", "four"])
```

```python
# 특정 요소가 있는 행만 선택
df[df["E"].isin(["two", "four"])]
```

```python
# 특정 컬럼 제거
del df["E"]
df
```

```python
# 각 컬럼 누적합
df.apply(np.cumsum)
```

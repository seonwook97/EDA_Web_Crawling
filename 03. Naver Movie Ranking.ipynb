````python
# requirements

import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup
````

## 2. 자동화를 위한 코드

```
https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date=20211001
```
- 날짜만 변경하면 우리가 원하는 기간 만큼 데이터를 얻을 수 있다

````python
date = pd.date_range('2021.01.01', periods=100, freq='D')
date
````

````python
date[0]
````

````python
date[0].strftime('%Y-%m-%d')
````

````python
date[0].strftime('%Y.%m.%d')
````

````python
# 문자열 format

test_string = "Hi, I'm {name}"
test_string.format(name='Zerobase')
test_string.format(name='Pinkwink')
````

````python
import time
from tqdm import tqdm

movie_date = []
movie_name = []
movie_point = []

for today in tqdm(date):
    url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date={date}'
    response = urlopen(url.format(date=today.strftime('%Y%m%d')))
    soup = BeautifulSoup(response, 'html.parser')
    
    end = len(soup.find_all('td', 'point'))
    
    movie_date.extend([today for n in range(0, end)])
    movie_name.extend([soup.select('div.tit5')[n].find('a').text for n in range(0, end)])
    movie_point.extend([soup.find_all('td', 'point')[n].string for n in range(0, end)])
    
    time.sleep(0.5)
````

````python
len(movie_date), len(movie_name), len(movie_point)
````

````python
movie_point[:5]
````

````python
movie_name[:5]
````

````python
movie = pd.DataFrame({
    'date': movie_date, 
    'name': movie_name,
    'point': movie_point
})
movie.tail()
````

````python
movie.info()
````

````python
movie['point'] = movie['point'].astype(float) 
movie.info()
````

## 3. 영화 평점 데이터 정리  

````python
import numpy as np
import pandas as pd
````

````python
movie = pd.read_csv('C:/Users/고선욱/Documents/ds_study/data/03. naver_movie_data.csv', index_col=0)
movie.tail()
````

- 영화 이름으로 인덱스를 잡는다
- 점수의 합산을 구한다
- 100일 간 네이버 영화 평점 합산 기준 베스트&워스트 10 선정

````python
# pivot table

movie_unique = pd.pivot_table(data=movie, index='name', aggfunc=np.sum)
movie_unique
````

````python
movie_best = movie_unique.sort_values(by='point', ascending=False) # 내림차순
movie_best.head()
````

````python
tmp = movie.query("name == ['화양연화']")
tmp
````

````python
# 시각화

import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='Malgun Gothic')
%matplotlib inline
# get_ipython().run_line_magic('matplotlib', 'inline')
````

````python
plt.figure(figsize=(20,8)) # x 20, y 8
plt.plot(tmp['date'], tmp['point']) # 선 그래프 x축 날짜, y축 평점 => 날짜에 따른 평점 변화를 선그래프로 표현(시계열)
plt.title('날짜별 평점')
plt.xlabel('날짜')
plt.ylabel('평점')
plt.xticks(rotation='vertical') # x축 범위를 세로로 표시
plt.legend(labels=['평점추이'], loc='best')
plt.grid(True)
plt.show()
````
